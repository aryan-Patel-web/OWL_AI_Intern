```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

print("Outliers in Fare:")
print("="*80)
fare_outliers = detect_outliers_iqr(df, 'Fare')
print(f"Number of outliers: {len(fare_outliers)}")
print(f"Percentage: {(len(fare_outliers)/len(df))*100:.2f}%")

print("\nOutliers in Age:")
print("="*80)
age_outliers = detect_outliers_iqr(df[df['Age'].notna()], 'Age')
print(f"Number of outliers: {len(age_outliers)}")
print(f"Percentage: {(len(age_outliers)/df['Age'].notna().sum())*100:.2f}%")
```

```python
summary_data = {
    'Feature': ['Overall', 'Male', 'Female', 'Class 1', 'Class 2', 'Class 3'],
    'Total_Passengers': [
        len(df),
        len(df[df['Sex'] == 'male']),
        len(df[df['Sex'] == 'female']),
        len(df[df['Pclass'] == 1]),
        len(df[df['Pclass'] == 2]),
        len(df[df['Pclass'] == 3])
    ],
    'Survived': [
        df['Survived'].sum(),
        df[df['Sex'] == 'male']['Survived'].sum(),
        df[df['Sex'] == 'female']['Survived'].sum(),
        df[df['Pclass'] == 1]['Survived'].sum(),
        df[df['Pclass'] == 2]['Survived'].sum(),
        df[df['Pclass'] == 3]['Survived'].sum()
    ]
}

summary_df = pd.DataFrame(summary_data)
summary_df['Survival_Rate_%'] = (summary_df['Survived'] / summary_df['Total_Passengers'] * 100).round(2)

print("Survival Summary:")
print("="*80)
print(summary_df)
```

```python
print("\n" + "="*80)
print("KEY INSIGHTS FROM EXPLORATORY DATA ANALYSIS")
print("="*80)

print("\n1. SURVIVAL RATE:")
survival_rate = (df['Survived'].sum() / len(df)) * 100
print(f"   - Overall survival rate: {survival_rate:.2f}%")
print(f"   - {df['Survived'].sum()} passengers survived out of {len(df)}")

print("\n2. GENDER ANALYSIS:")
female_survival = (df[df['Sex'] == 'female']['Survived'].sum() / len(df[df['Sex'] == 'female'])) * 100
male_survival = (df[df['Sex'] == 'male']['Survived'].sum() / len(df[df['Sex'] == 'male'])) * 100
print(f"   - Female survival rate: {female_survival:.2f}%")
print(f"   - Male survival rate: {male_survival:.2f}%")
print(f"   - Females were {female_survival/male_survival:.2f}x more likely to survive")

print("\n3. CLASS ANALYSIS:")
for pclass in [1, 2, 3]:
    class_survival = (df[df['Pclass'] == pclass]['Survived'].sum() / len(df[df['Pclass'] == pclass])) * 100
    print(f"   - Class {pclass} survival rate: {class_survival:.2f}%")

print("\n4. AGE ANALYSIS:")
print(f"   - Average age of survivors: {df[df['Survived'] == 1]['Age'].mean():.2f} years")
print(f"   - Average age of non-survivors: {df[df['Survived'] == 0]['Age'].mean():.2f} years")

print("\n5. MISSING DATA:")
print(f"   - Age: {df['Age'].isnull().sum()} missing ({(df['Age'].isnull().sum()/len(df))*100:.2f}%)")
print(f"   - Cabin: {df['Cabin'].isnull().sum()} missing ({(df['Cabin'].isnull().sum()/len(df))*100:.2f}%)")
print(f"   - Embarked: {df['Embarked'].isnull().sum()} missing")

print("\n6. FAMILY SIZE:")
avg_family_survivors = df[df['Survived'] == 1]['FamilySize'].mean()
avg_family_non_survivors = df[df['Survived'] == 0]['FamilySize'].mean()
print(f"   - Average family size of survivors: {avg_family_survivors:.2f}")
print(f"   - Average family size of non-survivors: {avg_family_non_survivors:.2f}")

print("\n" + "="*80)
print("EDA COMPLETED SUCCESSFULLY!")
print("="*80)
```
